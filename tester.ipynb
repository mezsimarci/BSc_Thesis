With this ipynb notebook you can easily play around with the simulator and the differential equation solver. There are a few examples prepared.

```python
import simulator
import diff_eq
import numpy as np
```

```python
gamma=[1000]     #the numbers of the different types of servers
N=sum(gamma)       #the total number of servers
B=10    #the limit of the queue lenghts of servers
lmbd=1.25    #the rate of arrival
mu=[[0]+[1,1.1,1.2,1.3,1.4]+[1.5 for i in range(6,B+1)]]   #the 0 in the beginning of each row makes sure that a demand cannot exit an empty queue
T1=10
T=30
T2=10
Tdiff=T1+T+T2
#the timespan
serverog=[]
for i in range(len(gamma)):
    serverog.append(np.zeros(gamma[i], dtype=int))     #the initial state of the servers, organised by type (initial queue lenght)
simulator.run(N=N,gamma=gamma,B=B,lmbd=lmbd,mu=mu,T=T,T1=T1,T2=T2,serverog=serverog,threshold=[3],d=5,plot=True,plotdensity=0.1,loadbalance="jiq",documentation=False,p=0.5,histogram=True,histbins=100,histmid=4)[0]
```
